#### 2.3<br>
#### Пошук медіани<br>
1) Найкращий випадок T(n) = O(n), коли ми виконаємо метод partition лише один раз.<br>
2) Найгірший випадок T(n) = T(n-1) + O(n) -> T(n) = O(n(n-1)) = O(n^2), коли ми знайдемо медіану перебравши всі інші значення.<br>

#### Пошук k-найменшого<br>
1) Найкращий випадок T(n) = O(n), коли ми виконаємо метод partition лише один раз.<br>
2) Найгірший випадок T(n) = T(n-1) + O(n) -> T(n) = O(n(n-1)) = O(n^2), коли ми знайдемо k-найменший <br>перебравши всі інші значення.<br>
#### Доведення за математиним сподіванням.
Нехай $N_1, N_2, \dots, N_k$ це підпроблеми, які ми маємо на $k$-кроці.<br>
$N_i$ крок успішний якщо $|N_i|<\frac{5}{6}|N_{i-1}|$.<br>
$Y_i$ - кількість рекурсивних викликів після $(i-1)$-ої успішної ітерації,<br>
$n_i$ - розмір підзадачі після $(i-1)$-ої успішної ітерації.<br>
Час виконання алгоритму $T - \sum_{k=0}^{i} n_k \cdot Y_k$.<br>
Нехай $n_i < \frac{5}{6}n_{i-1} < \frac{5}{6}n$.<br>
$Y_i$ - це випадкова змінна, що має геометричний розподіл з параметром $\frac{1}{2}$.
$E(Y_i) = 2$.<br>
$E\left(\sum_{k=0}^{i} n_k \cdot Y_k\right) = \sum_{k=0}^{i} \frac{5}{6}n \cdot Y_k = n\sum_{k=0}^{i}\frac{5}{6} E(Y{k})< n\sum_{k=0}^{i} \frac{5}{6}*2$ <an<br>
Де `a` - якась константа. Це означає, що б ми час quickSelect лінійний за математиним сподіванням.
#### Доведення математичною індукцією 
нехай n - кількість елементів в списку.
1) Якщо n = 1, то час виконаня T(1) = 0.<br>
2) Припустимо $T(k) \leq a*k$, наше індуктивне припущення. для k < n<br>
3) $T(n) \leq (розбиття) + (розділення)+ (виконання)$<br>
розбиття - (partition = O(n))<br>
розділення - (O(1))<br>
виконання - (O(a*n + b) = O(n))<br>
$T(n) \leq O(n) + O(1) + 2*O(n)$
$T(n) =  O(n)$<br>

### 3.2.

a) 
Найгірший випадок для quick sort, коли pivot розділяє список на частини n-1 і 1<br>
T(n) = T(n-1) + O(n) = O(n^2) (Master theorem)<br>

Найкращий випадок - коли pivot розділяє список на n/2 та n/2(дві рівні частини) <br> 
T(n) = 2T(n/2) + O(n) = O(nlogn) (Master theorem)<br>

б) Шукаючи кожен раз медіану, ми отримаєм додаткову складність.<br>
В кращому випадку - T(n) = T(n-k) + T(k) + O(n)+ O(n) = O(nlogn)  (Master theorem)
В гіршому випадку - T(n) = T(n-k) + T(k) + O(n^2)+ O(n) 

в) Шукаючи кожен раз медіану, ми отримаєм додаткову складність.(враховуючи математичне сподівання та індукцію)<br>
T(n) = T(n-k) + T(k) + O(n)+ O(n) = O(nlogn) (Master theorem)

г)Визначимо випадкову величину Xij, яка дорівнює кількості порівнянь елементів zi
та zj. Кількість порівнянь, які виконуються протягом роботи алгоритму, можна виразити
наступним чином: 
$\sum_{i=1}^{n-1} \sum_{j=1 + i}^{n} X_{ij}$

E($\sum_{i=1}^{n-1} \sum_{j=1 + i}^{n} X_{ij}$) = $\sum_{i=1}^{n-1} \sum_{j=1 + i}^{n} E(X_{ij})$   = P(Xi порівнюється з Xj)<br>
P{zi порівнюється з zj}= P{Першим опорним обрано zi або zj}
= P{Першим опорним обрано zi}+ P{Першим опорним обрано zj} = $\frac{1}{{j - i + 1}} + \frac{1}{{i - j + 1}} = \frac{2}{{j - i + 1}}$<br>

E(X) = $\sum_{i=1}^{n-1} \sum_{j=1 + i}^{n}  \frac{2}{{j - i + 1}}$ < 2*$\sum_{i=1}^{n-1} \sum_{k=1}^{n}  \frac{1}{k}$ < 2*$\sum_{i=1}^{n-1} O(lgn) = O(nlogn)$ <br>

```python

X = np.array([100, 600, 1100, 1600, 2100,2700,3300,3900]) 
sort_sorted = np.array([625, 663, 729, 790, 799,830,921,1012])
sort_opposite = np.array([649, 690, 694, 704, 731,820,875,956]) 
sort_random = np.array([617, 681, 660, 659, 667,701,750,769])

plt.plot(X, sort_sorted,  label='Sorted array') 
plt.plot(X, sort_opposite, label='Reversed sorted array')
plt.plot(X, sort_random, label='Random Sorted Array') 

plt.xlabel("N") 
plt.ylabel("Time in ns") 
plt.title("Quick Sort (Uniform pivot)") 
plt.legend() 
plt.show()
```
